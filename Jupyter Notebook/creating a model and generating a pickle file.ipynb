```python
import pandas as pd
```

```python
df1=pd.read_csv("/content/drive/MyDrive/Aegis/Machine Learning/Project/Dataset/train.csv")
df2=pd.read_csv("/content/drive/MyDrive/Aegis/Machine Learning/Project/Dataset/test.csv")
```

```python
train=df1.copy()
test=df2.copy()
```

```python
train.drop(['Unnamed: 0','id'],axis=1,inplace=True)
test.drop(['Unnamed: 0','id'],axis=1,inplace=True)
train.drop(['Arrival Delay in Minutes','Departure Delay in Minutes'],axis=1,inplace=True)
test.drop(['Arrival Delay in Minutes','Departure Delay in Minutes'],axis=1,inplace=True)
```

```python
#Here we are Encoding Male as 0 and Female as 1
train["Gender"]=train["Gender"].map({"Male":0,"Female":1})
test["Gender"]=test["Gender"].map({"Male":0,"Female":1})
```

```python
#Here We are Encoding disloyal customer as 0 and Loyal Customer as 1
train["Customer Type"]=train["Customer Type"].map({"disloyal Customer":0,"Loyal Customer":1})
test["Customer Type"]=test["Customer Type"].map({"disloyal Customer":0,"Loyal Customer":1})
```

```python
#Here, we are encoding Personal Travel as 0 and Business travel as 1
train["Type of Travel"]=train["Type of Travel"].map({"Personal Travel":0,"Business travel":1})
test["Type of Travel"]=test["Type of Travel"].map({"Personal Travel":0,"Business travel":1})
```

```python
#Here we are Encoding Economy as 0 ,Economy Plus 1 and Business class as 2
train["Class"]=train["Class"].map({"Eco":0,"Eco Plus":1,"Business":2})
test["Class"]=test["Class"].map({"Eco":0,"Eco Plus":1,"Business":2})
```

```python
X_train=train.iloc[:,:-1]
X_test=test.iloc[:,:-1]
y_train=train.iloc[:,-1]
y_test=test.iloc[:,-1]
```

```python
X_train
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.pipeline import Pipeline
pipe = Pipeline([('minmaxscale', MinMaxScaler()), ('RandomForest', RandomForestClassifier(bootstrap= False, criterion = 'entropy', n_estimators = 40))])
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred=pipe.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
accuracy_score(y_test,y_pred)
```
