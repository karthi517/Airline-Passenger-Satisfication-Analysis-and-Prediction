Here,I'm Creating a class and objects to for visulaization,because the if we are calling that any where we can view the visualization we no need to code it again and again

```python
#importing metrics to evaluatiing the dataset
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
#creating a class for doing visualization
class project:
  #creating a function for boxplot for checking outliers
  def outliers(self,df):
    for i in df:
      #giving size of the figure
      fig=plt.figure(figsize=(5,5))
      #plotting the boxplot
      sns.boxplot(data=df1,y=i)
  #Creating function for countplot to view the count of satisfied or not satified based on differenct paramenters
  def Countplot(self,value):
    for i in range(len(value)-1):
          #giving size of the figure
          fig=plt.figure(figsize=(5,5))
          #plotting the countplot
          sns.countplot(df1[value[i]],hue=df1[value[-1]])
          plt.title(value[i]+' '+value[-1])

  #Creating histogram to check how the data is skwed
  def histogram(self,value):
    for i in value:
      fig=plt.figure(figsize=(5,5))
      #Plotting histogram
      df1[i].hist()
      plt.xlabel(i)



  def evaluate(self,model, X_train, X_test, y_train, y_test):
    #predicting y_test and y_train
    y_test_pred = model.predict(X_test)
    y_train_pred = model.predict(X_train)

    print("TRAINIG RESULTS: \n===============================")
    #printing Confusion MAxtrix classification report and accuracy of training
    clf_report = pd.DataFrame(classification_report(y_train, y_train_pred, output_dict=True))
    print(f"CONFUSION MATRIX:\n{confusion_matrix(y_train, y_train_pred)}")
    print(f"ACCURACY SCORE:\n{accuracy_score(y_train, y_train_pred):.4f}")
    print(f"CLASSIFICATION REPORT:\n{clf_report}")

    #printing Confusion MAxtrix classification report and accuracy of testing
    print("TESTING RESULTS: \n===============================")
    clf_report = pd.DataFrame(classification_report(y_test, y_test_pred, output_dict=True))
    print(f"CONFUSION MATRIX:\n{confusion_matrix(y_test, y_test_pred)}")
    print(f"ACCURACY SCORE:\n{accuracy_score(y_test, y_test_pred):.4f}")
    print(f"CLASSIFICATION REPORT:\n{clf_report}")





```

#Preprocessing

Data Encoding is used to convert categorical values to integer we are having differnent encodore namely LabelEncoder,OneHotEncoder,mannual Encoding and etc,,,.

###Data Encoding

```python
#Here we are Encoding Male as 0 and Female as 1
df1["Gender"]=df1["Gender"].map({"Male":0,"Female":1})
df1_test["Gender"]=df1_test["Gender"].map({"Male":0,"Female":1})
```

```python
#Here We are Encoding disloyal customer as 0 and Loyal Customer as 1
df1["Customer Type"]=df1["Customer Type"].map({"disloyal Customer":0,"Loyal Customer":1})
df1_test["Customer Type"]=df1_test["Customer Type"].map({"disloyal Customer":0,"Loyal Customer":1})
```

```python
#Here, we are encoding Personal Travel as 0 and Business travel as 1
df1["Type of Travel"]=df1["Type of Travel"].map({"Personal Travel":0,"Business travel":1})
df1_test["Type of Travel"]=df1_test["Type of Travel"].map({"Personal Travel":0,"Business travel":1})
```

```python
#Here we are Encoding Economy as 0 ,Economy Plus 1 and Business class as 2
df1["Class"]=df1["Class"].map({"Eco":0,"Eco Plus":1,"Business":2})
df1_test["Class"]=df1_test["Class"].map({"Eco":0,"Eco Plus":1,"Business":2})
```
